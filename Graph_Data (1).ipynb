```python
import requests 
import pandas as pd
import numpy as np
from numpy.ma.core import true_divide
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from numpy.ma.core import true_divide
# your code here

url =  'https://www.alphavantage.co/query?function=TIME_SERIES_Monthly_Adjusted&symbol=IBM&apikey=0TOJPG6CWZ8MLZ5E'
r = requests.get(url)
data = r.json()
prices = pd.DataFrame.from_dict(data['Monthly Adjusted Time Series'], orient='index'
)
prices.reset_index(inplace=True)
prices.rename(columns={'1. open': 'Open','2. high': 'High','3. low': 'Low', '4. close': 'Close', '5. volume': 'Volume', 'index': 'Date'})

index = pd.date_range(start = '2022-03-01',end='2022-03-07')
```

```python
prices["2. high"]=prices["2. high"].astype(float)
prices["3. low"]=prices["3. low"].astype(float)

```

```python

prices.plot(x="index", y=["2. high", "3. low"], kind="bar")
plt.xlabel('IBM Monthly Stock Price')
plt.ylabel('Time') #I was going to do the daily high and lows for the most recent week but I couldnt get the date time to work countless times.
plt.title('Daily High/Low 3/20/22-3/27/22 $IBM')
```
