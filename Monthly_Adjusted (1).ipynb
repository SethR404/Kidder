```python
import requests 
import pandas as pd
import numpy as np
from numpy.ma.core import true_divide
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def stocks(ticker=True):
  url = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY_ADJUSTED&symbol='+ticker+'&apikey=0TOJPG6CWZ8MLZ5E'
  r = requests.get(url)
  data = r.json()
  stock = pd.DataFrame.from_dict(data['Monthly Adjusted Time Series'], orient='index')
  stock.reset_index(inplace=True)
  stock.rename(columns={'1. open': 'Open','2. high': 'high','3. low': 'Low', '4. close': 'Close', '5. adjusted close': 'Adjusted Close', '6. volume': 'Volume', '7. dividend amount': 'Dividend Amount', 'index': 'Date'},inplace=True)
  stock['ticker']=ticker
  return stock
```

```python
stocks(ticker='IBM')
```
