```python
import requests 
import pandas as pd
import numpy as np
from numpy.ma.core import true_divide
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def quarter(month=True):
  temp=month/3
  if temp<=1:
    return 1
  elif temp>1 and temp<=2:
    return 2
  elif temp>2 and temp<=3:
    return 4
  else:
      return 4
```

```python
 ticker='IBM'
 def Stocks(ticker=True):
  url =  'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY_ADJUSTED&symbol=IBM&apikey=0TOJPG6CWZ8MLZ5E'
  r = requests.get(url)
  data = r.json()
  stock=pd.DataFrame.from_dict(data['Monthly Adjusted Time Series']).T
  stock.reset_index(inplace=True)
  stock.rename(columns={'1. open': 'Open','2. high': 'High','3. low': 'Low', '4. close': 'Close', '5. adjusted close': 'Adjusted Close', '6. volume': 'Volume', '7. dividend amount': 'Dividend Amounnt', 'index': 'date'},inplace=True)
  stock['date'] = pd.to_datetime(stock['date'])
  stock['year'] = stock['date'].dt.year
  stock['month'] = stock['date'].dt.month
  columns=stock.columns

  for i in columns[1:8]:
    stock[i]=stock[i].astype(float)
  stock['quarter'] = stock.apply(lambda x: quarter(x['month']),axis=1)
  stock=stock.groupby(['year', 'quarter', 'date'])['Open'].mean().reset_index()
  stock['ticker']=ticker
  return stock

A=Stocks(ticker='A')
IBM=Stocks(ticker='IBM')

pd.concat([IBM,A])

stock_list=['IBM','A']
temp=pd.DataFrame()
for j in stock_list:
  T=Stocks(ticker=j)
  temp=pd.concat([temp,T])
temp

```
